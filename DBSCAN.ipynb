```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles,make_moons
```

```python
X,y = make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(X,y)
```

```python
centres = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(centres[:,0],centres[:,1],marker='*',edgecolors='black')
plt.show()
```

# DBSCAN
- Density Based Spatial Clustering of Application with Noise
- Based upon the idea, that a cluster is a high density area surrounded by  low density area.
- Starts by exploring a small area, if density is 'good enough' ir considered a part of cluster, and explores neighbour and so on. So it groups region on the basis of density!

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.2,min_samples=5)
```

```python
ypred = dbs.fit_predict(X)
```

```python
print(ypred)
```

```python
plt.scatter(X[:,0],X[:,1],c=ypred)
plt.show()
```
